##### Scipy:
##### We have the min and max temperatures in a city In India for each months of the year.
##### We would like to find a function to describe this and show it graphically, the dataset given below.
##### Task:
##### 1. fitting it to the periodic function
##### 2. plot the fit
##### Data
##### Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
##### Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, 'ro')
plt.plot(months, temp_min, 'bo')
plt.xlabel('Month')
plt.ylabel('Min and max temperature')
```

##### Fitting it to Periodic function

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      temp_max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      temp_min, [-40, 20, 0])
```

##### Plotting the fit

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, temp_max, 'go')
plt.plot(days, yearly_temps(days, *res_max), 'm-')
plt.plot(months, temp_min, 'co')
plt.plot(days, yearly_temps(days, *res_min), 'y-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read titanic dataset
titanic=pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

```python
titanic.head()
```

```python
pd_ser=titanic['sex']
wedges = ['Male','Female']
fig1, ax1 = plt.subplots()
prop=[pd_ser[pd_ser=='male'].size,pd_ser[pd_ser=='female'].size]
ax1.pie(prop, labels=wedges, autopct='%1.2f%%')
plt.show()
```

```python
male_s=titanic[titanic['sex']=='male']
female_s=titanic[titanic['sex']=='female']

fig, ax = plt.subplots()
m=ax.scatter('age','fare',color='red',data=male_s, alpha=0.7)
n=ax.scatter('age','fare',color='blue',data=female_s, alpha=0.7)
ax.set_xlabel('age')
ax.set_ylabel('fare')

ax.legend((m,n),('male','female'))

plt.show()
```
